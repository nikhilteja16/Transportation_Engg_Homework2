```python
dxf = pd.read_csv('bike_data.csv')
```

```python
dxf
```

```python
import statsmodels.api as sm
from patsy import dmatrices
y = dxf['TotalBikesCount']
formula = 'TotalBikesCount ~ C(Mon)+C(Tue)+C(Wed)+C(Thu)+C(Fri)+C(Sat)+C(Sun)+C(holiday)+DayLightHrs+AvgTempInC+PRCP_IN+C(Dry_Day)'
y,x = dmatrices(formula,dxf,return_type = 'dataframe')
results = sm.GLM(y,x,family = sm.families.Poisson()).fit().summary()
```

```python
print(results)
```

```python
#All variables are signifianct here as P=0.000 for all of them.
# On NON-holidays ssuch as mon tue wed thursday friday sat and sun the coefficients are positive meaning
# that on these days the bike count is influenced linearly  and the bike count increases as the non holidays
# increase
```

```python
# The HOLIDAYS have a negative coefficient and hence would influence the bike count inversely. This means
# the model would predict that as the number of holidays increase the bike counts would reduce.
```

```python
# DayLightHrs and AvgTempInc also influence the bike counts positively. As the day light increases and the Avg
#Temp increases the model showcases that the bike counts in the city would also increase. This is because of the 
#positive coefficients of both DayLightHrs and AvgTempInc.

# PRCP_IN or Precipitation in inches influences the bike counts inversely or as the precipitation increases
# the model predicts that the bike counts in the city would reduce.
# Hence Both the holiday and PRCP_IN have a negative influence on the bike counts
```
